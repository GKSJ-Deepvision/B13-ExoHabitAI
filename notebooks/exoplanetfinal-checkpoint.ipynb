```python
import pandas as pd

df = pd.read_csv("exohabit_clean.csv", comment='#')
df.head()
```

```python
required_columns = [
    'pl_rade',      # Planet Radius
    'pl_bmasse',    # Planet Mass
    'pl_orbper',    # Orbital Period
    'pl_orbsmax',   # Semi-major axis
    'pl_eqt',       # Equilibrium Temperature
    'pl_dens',      # Planet Density
    'st_teff',      # Star Temperature
    'st_lum',       # Star Luminosity
    'st_met',       # Star Metallicity
    'st_spectype'   # Star Type
]

df = df[required_columns]
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(how='all')
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
df['st_spectype'] = df['st_spectype'].fillna(df['st_spectype'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df = df[df['pl_rade'] > 0]
df = df[df['pl_bmasse'] > 0]
df = df[df['pl_eqt'] > 0]
```

```python
from scipy import stats
import numpy as np

numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

z_scores = np.abs(stats.zscore(df[numeric_cols]))
df = df[(z_scores < 3).all(axis=1)]
```

```python
df['habitability_score'] = (
    (1 / abs(df['pl_eqt'] - 288)) +
    (1 / abs(df['pl_rade'] - 1)) +
    (1 / abs(df['pl_orbsmax'] - 1)) +
    (1 / abs(df['st_lum'] - 1))
)
```

```python
df['orbital_stability'] = df['pl_orbsmax'] / df['pl_orbper']
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['st_spectype'])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
```

```python
from sklearn.preprocessing import StandardScaler

# Select only continuous numeric columns (NOT dummy columns)
scale_columns = [
    'pl_rade',
    'pl_bmasse',
    'pl_orbper',
    'pl_orbsmax',
    'pl_eqt',
    'pl_dens',
    'st_teff',
    'st_lum',
    'st_met',
    'habitability_score',
    'orbital_stability'
]

scaler = StandardScaler()
df[scale_columns] = scaler.fit_transform(df[scale_columns])
```

```python
df[scale_columns].dtypes
```

```python
import numpy as np

np.isinf(df[scale_columns]).sum()
```

```python
df[scale_columns] = df[scale_columns].replace([np.inf, -np.inf], np.nan)
```

```python
for col in scale_columns:
    df[col] = df[col].fillna(df[col].median())
```

```python
np.isinf(df[scale_columns]).sum()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[scale_columns] = scaler.fit_transform(df[scale_columns])
```

```python
df['target'] = np.where(
    df['habitability_score'] > df['habitability_score'].median(),
    1,
    0
)
```

```python
df.to_csv("preprocessed.csv", index=False)
```
