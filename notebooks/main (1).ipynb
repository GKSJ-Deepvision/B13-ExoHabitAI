```python
expected_features = {
    'pl_name': 'pl_name',           # Planet Name (ID)
    'hostname': 'hostname',         # Host Star Name (ID)
    'pl_rade': 'pl_rade',           # Planet Radius [Earth Radii]
    'pl_radj': 'pl_radj',           # Planet Radius [Jupiter Radii] - For Recovery
    'pl_masse': 'pl_masse',         # Planet Mass [Earth Masses]
    'pl_massj': 'pl_massj',         # Planet Mass [Jupiter Masses] - For Recovery
    'pl_orbper': 'pl_orbper',       # Orbital Period [Days]
    'pl_orbsmax': 'pl_orbsmax',     # Semi-Major Axis [AU]
    'pl_eqt': 'pl_eqt',             # Equilibrium Temp [Kelvin]
    'pl_dens': 'pl_dens',           # Density [g/cm**3]
    'st_teff': 'st_teff',           # Star Temp [Kelvin]
    'st_lum': 'st_lum',             # Star Luminosity [log(Solar)]
    'st_met': 'st_met',             # Star Metallicity [dex]
    'st_spectype': 'st_spectype'    # Star Type (Spectral)
}
```

```python
# Select available columns
cols_to_use = [col for col in expected_features.values() if col in df.columns]
processed_df = df[cols_to_use].copy()
```

```python
# Ensure numeric types
numeric_cols = ['pl_rade', 'pl_radj', 'pl_masse', 'pl_massj', 'pl_orbper', 
                'pl_orbsmax', 'pl_eqt', 'pl_dens', 'st_teff', 'st_lum', 'st_met']
for col in numeric_cols:
    if col in processed_df.columns:
        processed_df[col] = pd.to_numeric(processed_df[col], errors='coerce')
        # Replace 0 with NaN so we can impute/recover it later
        processed_df[col] = processed_df[col].replace(0, np.nan)

print(f"Data Loaded. Shape: {processed_df.shape}")
```

```python
# 1. Missing Value Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(processed_df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Value Heatmap (Before Cleaning)')
plt.show()
```

```python
# 2. Summary Statistics
print("\nSummary Statistics (Raw):")
print(processed_df.describe().T[['count', 'mean', 'min', 'max']])
```

```python

print("\nDUPLICATE Planet REMOVAL")

# 1. Count nulls in each row (to identify the "best" row for each planet)
# We want the row that has the most  values for our important columns
search_cols = ['pl_rade', 'pl_masse', 'pl_orbper', 'pl_eqt', 'st_teff']
processed_df['null_count'] = processed_df[search_cols].isnull().sum(axis=1)

# 2. Sort: Put the rows with FEWEST nulls at the top
processed_df = processed_df.sort_values(by=['pl_name', 'null_count'], ascending=[True, True])

# 3. Drop Duplicates
# Now, 'keep=first' will automatically keep the best row because we sorted them to the top.
initial_count = len(processed_df)
processed_df = processed_df.drop_duplicates(subset=['pl_name'], keep='first')
final_count = len(processed_df)

# Remove the helper column
processed_df = processed_df.drop(columns=['null_count'])
print(processed_df.shape)

print(f"Raw Observations: {initial_count}")
print(f"Unique Planets:   {final_count}")
print(f"Removed {initial_count - final_count} duplicate observations.")
print("Note: This matches the real-world number of confirmed exoplanets (~5600-6000).")
```

HANDLING MISSING DATA (With Unit Conversion)

```python
# (Jupiter -> Earth Conversion)

# Instead of imputing the median, we use Jupiter units if Earth units are missing.

# Radius: 1 Jupiter Radius = 11.209 Earth Radii
if 'pl_radj' in processed_df.columns:
    mask = processed_df['pl_rade'].isnull() & processed_df['pl_radj'].notnull()
    processed_df.loc[mask, 'pl_rade'] = processed_df.loc[mask, 'pl_radj'] * 11.209
    print(f"  - Recovered {mask.sum()} radii using Jupiter conversion.")

# Mass: 1 Jupiter Mass = 317.8 Earth Masses
if 'pl_massj' in processed_df.columns:
    mask = processed_df['pl_masse'].isnull() & processed_df['pl_massj'].notnull()
    processed_df.loc[mask, 'pl_masse'] = processed_df.loc[mask, 'pl_massj'] * 317.8
    print(f"  - Recovered {mask.sum()} masses using Jupiter conversion.")
```

```python

# IMPUTATION (Median/Mode)

# Numerical -> Median
impute_cols = ['pl_rade', 'pl_masse', 'pl_orbper', 'pl_orbsmax', 'pl_eqt', 'pl_dens', 
               'st_teff', 'st_lum', 'st_met']
for col in impute_cols:
    if col in processed_df.columns:
        median_val = processed_df[col].median()
        processed_df[col] = processed_df[col].fillna(median_val)

# Categorical -> Mode
if 'st_spectype' in processed_df.columns:
    mode_val = processed_df['st_spectype'].mode()[0]
    processed_df['st_spectype'] = processed_df['st_spectype'].fillna(mode_val)

print(f"Remaining Missing Values: {processed_df.isnull().sum().sum()}")
```

```python
#OUTLIER DETECTION 

# We only remove values that are physically impossible in our universe.
# We DO NOT cap large values because massive planets are real and important.

initial_rows = len(processed_df)

# 1. Radius must be positive
processed_df = processed_df[processed_df['pl_rade'] > 0]

# 2. Temperature must be positive (Absolute Zero is 0 K)
processed_df = processed_df[processed_df['pl_eqt'] > 0]

# 3. Orbital Period must be positive
if 'pl_orbper' in processed_df.columns:
    processed_df = processed_df[processed_df['pl_orbper'] > 0]

# 4. Mass must be positive
if 'pl_masse' in processed_df.columns:
    processed_df = processed_df[processed_df['pl_masse'] > 0]

dropped_count = initial_rows - len(processed_df)
print(f"Removed {dropped_count} rows with physically impossible values (negatives/zeros).")
print(f"Remaining Planets: {len(processed_df)}")
print("Skipping Statistical Capping (IQR/Z-Score) to preserve real massive planets.")
```

```python
# FEATURE ENGINEERING

# Habitability Score Index (Simplified ESI)

# Weighted distance from Earth values (Radius=1, Temp=288K)

# Formula: Geometric Mean of (1 - abs(diff))
w_rad = 0.57
w_temp = 1.07
esi_rad = (1 - abs((processed_df['pl_rade'] - 1) / (processed_df['pl_rade'] + 1))) ** w_rad
esi_temp = (1 - abs((processed_df['pl_eqt'] - 288) / (processed_df['pl_eqt'] + 288))) ** w_temp
processed_df['habitability_score'] = np.sqrt(esi_rad * esi_temp)

# Stellar Compatibility Index
# Clean Star Type first (Keep only O, B, A, F, G, K, M)

processed_df['star_class'] = processed_df['st_spectype'].astype(str).str[0].str.upper()
valid_stars = ['O', 'B', 'A', 'F', 'G', 'K', 'M']
processed_df.loc[~processed_df['star_class'].isin(valid_stars), 'star_class'] = 'Unknown'

# Assign scores (G/K best, M okay, O/B bad)
star_scores = {'G': 1.0, 'K': 0.9, 'F': 0.7, 'M': 0.5, 'A': 0.3, 'B': 0.1, 'O': 0.0, 'Unknown': 0.5}
processed_df['stellar_compatibility'] = processed_df['star_class'].map(star_scores)

# 3. Orbital Stability Factor
# Log relationship between Period and Semi-Major Axis
processed_df['orbital_stability'] = np.log1p(processed_df['pl_orbper'] * processed_df['pl_orbsmax'])
```

```python
if 'star_class' not in processed_df.columns:
    print("Re-creating missing 'star_class' column...")
    
    # Ensure we have the source column
    if 'st_spectype' in processed_df.columns:
        # Extract first letter (G, K, M...) and uppercase it
        processed_df['star_class'] = processed_df['st_spectype'].astype(str).str[0].str.upper()
        
        # Filter to only keep valid types (O, B, A, F, G, K, M)
        valid_stars = ['O', 'B', 'A', 'F', 'G', 'K', 'M']
        processed_df.loc[~processed_df['star_class'].isin(valid_stars), 'star_class'] = 'Unknown'
    else:
        print("ERROR: 'star_spectype' column is also missing! Check Section 5 (Renaming).")
```

```python
mapping = {
    'pl_name': 'planet_name',
    'hostname': 'host_star_name',
    'pl_rade': 'radius_earth',
    'pl_masse': 'mass_earth',
    'pl_orbper': 'orbital_period',
    'pl_orbsmax': 'semimajor_axis',
    'pl_eqt': 'eq_temp_k',
    'pl_dens': 'density',
    'st_teff': 'star_temp_k',
    'st_lum': 'star_luminosity',
    'st_met': 'star_metallicity',
    'st_spectype': 'star_spectype'
}

# 2. Rename columns if they exist
processed_df = processed_df.rename(columns=mapping)
```

```python

# CATEGORICAL ENCODING & SCALING
from sklearn.preprocessing import StandardScaler, RobustScaler

# 1. One-Hot Encoding for Star Class
processed_df = pd.get_dummies(processed_df, columns=['star_class'], prefix='star')

# 2. Feature Scaling
# Select numerical columns for scaling
#  Scaling (Use RobustScaler)
# Since we have "Monster" planets, RobustScaler is better than StandardScaler.
scale_cols = ['radius_earth', 'mass_earth', 'orbital_period', 'semimajor_axis', 
              'eq_temp_k', 'density', 'star_temp_k', 'star_luminosity', 'star_metallicity']

scaler = RobustScaler() 
scaled_data = scaler.fit_transform(processed_df[scale_cols])
scaled_df = pd.DataFrame(scaled_data, columns=[f"{c}_scaled" for c in scale_cols], index=processed_df.index)
processed_df = pd.concat([processed_df, scaled_df], axis=1)
```

```python

# TARGET VARIABLE CREATION
# Rule-based classification for training labels
# Criteria: Rocky (0.5-1.6 Earth Radii) AND Habitable Temp (200-330 K)

# Target
processed_df['habitable_binary'] = 0
mask_habitable = (
    (processed_df['radius_earth'] >= 0.5) & (processed_df['radius_earth'] <= 1.6) &
    (processed_df['eq_temp_k'] >= 200) & (processed_df['eq_temp_k'] <= 330)
)
processed_df.loc[mask_habitable, 'habitable_binary'] = 1

# Multi-class (0=No, 1=Optimistic, 2=Conservative)
# (Optional simplified version)
#processed_df['habitable_class'] = processed_df['habitable_binary']

print("Class Distribution:")
print(processed_df['habitable_binary'].value_counts())
```
